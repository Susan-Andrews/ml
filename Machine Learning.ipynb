# Supervised

Labelled data is given as input.Labelled data means some input data is already tagged with correct output .

classified into two * regression (output is continuous) weather forecasting * classification (any two output only like true or false male or female ...)

Some of the popular regression algorithms are:
* Linear 
* Regression trees
* Non -linear
* Bayseon linear
* Polynomial

Popular categories of Classification algorithms are:
* Logistic Regression 
* Random Forest
* Decision trees
* SVM


Aim of supervised learning is to find a mapping function to map input variable (x) with output variable (y) : x->f(y)
Advantages:
* predict output
* exact idea about classes of objects since labelled input solves real world problem.
    
Disadvantages:
* Not suitable for complex tasks
* cant predict output if test data is different from training dataset .
